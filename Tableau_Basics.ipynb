1.Explain what Tableau is and why it is preferred over traditional spreadsheet tools for data analysis.

Also mention any two industries where Tableau is commonly used and explain one use case.


Ans-

Tableau is a powerful data visualization and business intelligence tool used to

connect, analyze, and present data in the form of interactive dashboards and

reports. It allows users to turn raw data into meaningful insights without

needing advanced programming skills.

Why Tableau is preferred over traditional spreadsheet tools (like Excel)

1.Advanced Data Visualization

Tableau creates interactive and dynamic visuals (charts, maps, dashboards).

Spreadsheets mostly rely on static charts.

2.Handles Large Data Efficiently

Tableau can process millions of rows of data quickly.

Excel slows down or crashes with large datasets.

3.Drag-and-Drop Interface

Easy to use: no need for complex formulas.

In Excel, you often need formulas like VLOOKUP, Pivot Tables, etc.

4.Real-Time Data Connection

Tableau connects directly to databases (SQL, cloud, etc.).

Excel usually works with imported/static data.

5.Interactive Dashboards

Users can apply filters, drill down, and explore data visually.

Excel dashboards are less interactive.

6.Better Data Storytelling

Tableau helps present insights clearly using dashboards and stories.

Excel is more focused on data storage and calculations.

2.Retail

-Make faster decisions

9. What is Tableau Prep, and why is it important in real-world data projects?

What is Tableau Prep, and why is it important in real-world data projects?

Then explain how you would:

Remove null values

Remove duplicate records

Change incorrect data types (Answer in steps, no tool access required.)

Ans-


*Tableau Prep is a data preparation tool used to clean, combine, and transform

raw data before using it in Tableau for analysis and visualization.

It helps convert messy, unstructured data into clean, analysis-ready data

Why is it Important in Real-World Projects?

1.Data Cleaning

Removes errors like nulls, duplicates, incorrect formats

2.Data Transformation

Change data types, split/merge columns, create new fields

3.Combine Multiple Data Sources

Join, union, or aggregate data from different files/databases

4.Improves Accuracy

Clean data = reliable insights and better decisions

5.Saves Time

Automates repetitive data cleaning tasks using flows

How to Perform Common Tasks in Tableau Prep

1.Remove Null Values

Steps:

-Open your dataset in Tableau Prep

-Go to the Profile Pane

-Identify fields with null values

-Click on “Null” in the field

-Select:

Filter Data → Exclude (to remove null rows)

Alternative:

Add a Filter step

Set condition: Field ≠ Null

2.Remove Duplicate Records

Steps:

-Add a Clean Step

-Select the columns that define duplicates (e.g., Order ID, Customer Name)

-Click on:

More options → Remove Duplicate Rows

Alternative:

-Use Aggregate Step

-Group by unique fields to eliminate duplicates

3. Change Incorrect Data Types

Steps:

-Go to the Profile Pane

-Find the column with wrong data type

-Click on the data type icon (e.g., ABC, #, calendar)

-Change to required type:

String → Number

String → Date

Number → Date (if applicable)

Example:

Convert Order Date (string) → Date format

In a simpler words

Tableau Prep is a data preparation tool used to clean, transform, and combine

data before analysis.

It is important in real-world projects because it ensures data quality and accuracy.

-To remove null values, filter them out in the profile pane.

-To remove duplicates, use a clean step or aggregate step.

-To correct data types, change them directly in the profile pane based on the required format.
